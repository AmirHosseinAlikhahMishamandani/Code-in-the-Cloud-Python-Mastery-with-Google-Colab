## If Statement

```python
x = 10
if x > 5:
    print("x is greater than 5")
```

### If-Else Statement

```python
x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
```

### Elif Statement

```python
x = 3
if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is less than 5")
```

### Nested If Statement

```python
score = 55

if score >= 60:
    print("Congratulations! You have passed the course.")
else:
    print("Unfortunately, you did not pass the course.")
    if 50 <= score < 60:
        print("You have the option to retake the exam.")
    else:
        print("Your score is too low to retake the exam.")
```

### One-Line If Statement



```python
x = 10
if x > 5: print("x is greater than 5")
```

### Ternary Operator

```python
x = 10
result = "x is greater than 5" if x > 5 else "x is not greater than 5"
print(result)
```

### Comparisons without if

```python
age = 18
print(age == 18)
```

### Membership Operators

```python
my_list = [1, 2, 3, 4, 5]
print(3 in my_list)
print(6 not in my_list)
```

###  Identity Operators

```python
x = [1, 2, 3]
y = [1, 2, 3]
z = x
print(x is z)
print(x is y)
print(x is not y)
```

## Project: Driver's License Eligibility Checker

```python
name = input("Please enter your name: ")
age = int(input("Please enter your age: "))
written_test_passed = input("Did you pass the written test? (yes/no): ").lower() == 'yes'

parental_consent = False
if age < 18:
    parental_consent_input = input("Do you have parental consent? (yes/no): ").lower()
    parental_consent = parental_consent_input == 'yes'

if age >= 18:
    print("You are eligible for a driver's license.")
    if written_test_passed:
        print("You have passed the written test.")
    else:
        print("You need to pass the written test to proceed.")
elif age < 18:
    if parental_consent:
        print("You are under 18, but you have parental consent.")
        if written_test_passed:
            print("You have passed the written test.")
            print("You are eligible for a driver's license.")
        else:
            print("You need to pass the written test to proceed.")
    else:
        print("You are under 18 and require parental consent to proceed.")
else:
    print("Invalid age.")
```

## while Loop

### while Loop

```python
i = 0
while i < 5:
  print(i)
  i += 1
```

### while loop with break

```python
while True:
    response = input("Enter 'quit' to exit: ")
    if response == 'quit':
        break
    else:
        print("You entered:", response)
```

### while loop with flag

```python
active = True
while active:
    response = input("Enter 'quit' to exit: ")
    if response == 'quit':
        active = False
    else:
        print("You entered:", response)
```

### while loop with continue

```python
number = 0
while number <= 10:
    number += 1
    if number % 3 == 0:
        continue
    print(number)
```

### Infinite while loop

```python
number = 0
while number < 10:
  print(number)
  # number += 1
```
